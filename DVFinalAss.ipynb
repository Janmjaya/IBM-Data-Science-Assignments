```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

```python
df_dss = pd.read_csv('https://cocl.us/datasciece_survey_data/Data_Science_Topics_Survey.csv')
print('Data downloaded and read into a dataframe!')
```

```python
df_dss.head()
```

```python
pos1 = 1
colname1 = df_dss.columns[pos1]

pos2 = 2
colname2 = df_dss.columns[pos2]

pos3 = 3
colname3 = df_dss.columns[pos3]

pos4 = 4
colname4 = df_dss.columns[pos4]

pos5 = 5
colname5 = df_dss.columns[pos5]

pos6 = 6
colname6 = df_dss.columns[pos6]

df_dss.rename(columns={colname1:'Data Visualization'}, inplace=True)
df_dss.rename(columns={colname2:'Machine Learning'}, inplace=True)
df_dss.rename(columns={colname3:'Data Analysis / Statistics'}, inplace=True)
df_dss.rename(columns={colname4:'Big Data (Spark / Hadoop)'}, inplace=True)
df_dss.rename(columns={colname5:'Data Journalism'}, inplace=True)
df_dss.rename(columns={colname6:'Deep Learning'}, inplace=True)
```

```python
df_dss.head()
```

```python
df_dss.drop(['Timestamp'], axis=1, inplace=True)
df_dss.head(2)
```

```python
df_dssT = df_dss.T
```

```python
df_dssT
```

```python
df_dssT.sort_index(inplace=True)
```

```python
df_dssT
```

```python
df_dssT_list = []
for search_string in ['Not interested','Somewhat interested', 'Very interested']:
    #use above method but rename the series instead of setting to
    # a columns. The append to a list.
    df_dssT_list.append(df_dssT.apply(lambda x: x.str.contains(search_string))
                     .sum(axis=1) ## HERE IS SUM in place of any
                     .astype(int)
                     .rename(search_string))

#concatenate the list of series into a DataFrame with the original df
df_dssT = pd.concat([df_dssT] + df_dssT_list, axis=1)
df_dssT
```

```python
df2 = df_dssT.loc[:, ['Not interested','Somewhat interested', 'Very interested']]
df2
```

```python
df2 = df_dssT.loc[:, ['Very interested','Somewhat interested','Not interested']]
df2
```

```python
df3 = df2.sort_values("Very interested", ascending=[False])
df3
```

```python
df3['Total'] = df3.sum(axis=1)
# let's view the first five elements and see how the dataframe was changed
df3.head()
```

```python
df3.sort_values("Very interested", ascending=[False])
```

```python
df3['Very interested Per'] = df3['Very interested']/df3['Total']
df3['Somewhat interested Per'] = df3['Somewhat interested']/df3['Total']
df3['Not interested Per'] = df3['Not interested']/df3['Total']
df3.head()
```

```python
df3['Very interested Per'] = df3['Very interested Per'] * 100
df3['Somewhat interested Per'] = df3['Somewhat interested Per'] * 100
df3['Not interested Per'] = df3['Not interested Per'] * 100
df3.head()
```

```python
df4 = df3.loc[:, ['Very interested Per','Somewhat interested Per','Not interested Per']]
df4
```

```python
df4['Very interested'] = df4['Very interested Per']
df4['Somewhat interested'] = df4['Somewhat interested Per'] 
df4['Not interested'] = df4['Not interested Per']
df4.head()
```

```python
df5 = df4.loc[:, ['Very interested','Somewhat interested','Not interested']]
df5
```

```python
df6 = df5.round(2)
```
